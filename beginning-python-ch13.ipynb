```python
import sqlite3

def convert(value):
    if value.startswith('~'):
        return value.strip('~')
    if not value:
        value = '0'
    if value == '\n':
        value = '0'
    return float(value)

conn = sqlite3.connect('food.db')
curs = conn.cursor()

curs.execute('''
CREATE TABLE food (
id TEXT PRIMARY KEY,
FdGrp_Cd TEXT,
Long_Desc TEXT,
Shrt_Desc TEXT,
ComName TEXT,
ManufacName TEXT,
Survey BOOL,
Ref_desc TEXT,
Refuse FLOAT,
SciName TEXT,
N_Factor FLOAT,
Pro_Factor FLOAT,
Fat_Factor FLOAT,
CHO_Factor FLOAT
)
''')

query = 'INSERT INTO food VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?,?)'
field_count = 14

for line in open('SR-Leg_ASC/FOOD_DES.txt', encoding='latin-1'):
    fields = line.split('^')
    vals = [convert(f) for f in fields[:field_count]]
    curs.execute(query, vals)

conn.commit()
conn.close()
```

```python
import sqlite3, sys

conn = sqlite3.connect('food.db')
curs = conn.cursor()

query = 'SELECT * FROM food WHERE Refuse >= 75 ORDER BY Pro_Factor'
print(query)
curs.execute(query)
names = [f[0] for f in curs.description]
for row in curs.fetchall():
    for pair in zip(names, row):
        print('{}: {}'.format(*pair))
    print()
```
